```python
# Tuples
# Tuples are an ordered sequence
# Tuples are represented by parentheses
# Tuples are immutable
Ratings = (10, 9, 6, 5, 10, 8, 9, 6, 2)
print(Ratings)

myTuple = ("John", 2, 33.7)
print(myTuple)

# Combining tuples
tuple2 = myTuple + ("Ade", 76)
print(tuple2)

#Length of a tuple
print(len(tuple2))

# Accessing elements within a tuple
print(myTuple[0])

print(tuple2[1:4])

# Check Data Type
print(type (myTuple))

print(type (23))

# Tuples are Immutable
numbers = (10, 9, 6, 5, 10, 8, 9, 6, 2)
# numbers[2] = 77 # Type Error
print(numbers)

# Sorting Tuples
# To sort tuples, we used the sorted() method
NumbersSorted = sorted(numbers)
print(NumbersSorted)

# Accessing Nested Tuples
Note = (1, 2, ("Max Williams", 76, ("rock", "smart")), "bright", ("disco", 55, (3, 4)))
print(Note[2][0])
print(Note[2][2][1])

# Count method
# The count() method for a tuple is used to count how many times a specified element appears in the tuple.
fruits = ("apple", "banana", "apple", "orange")
print(fruits.count("apple")) #Counts the number of times apple is found in tuple.
#Output: 2


# index method
# The index() method in a tuple is used to find the first occurrence of a specified value and returns its position (index). If the value is not found, it raises a ValueError.
# Syntax: tuple.index(value)

fruits = ("apple", "banana", "orange","apple")
print(fruits.index("apple")) #Returns the index value at which apple is present.
#Output: 0


# Sum method
# The sum() function in Python can be used to calculate the sum of all elements in a tuple, provided that the elements are numeric (integers or floats).
# syntax: sum(tuple)

numbers = (10, 20, 5, 30)
print(sum(numbers))
#Output: 65


# min() and max() method
# Find the smallest (min()) or largest (max()) element in a tuple.

numbers = (10, 20, 5, 30)
print(min(numbers)) 
#Output: 5
print(max(numbers))
#Output: 30





```

```python
# List (Arrays)
# Lists are also an ordered sequence
# Lists are represented by square brackets
# Lists are mutable

first_list = ["Michael Sax", 23, 44.77, True, False, None, {"name": "John"}, (1, 2, "99")]
print(first_list)


# Combining Lists usind the plus operator
# Doesn't modify the original list
# Returns a new list

second_list = first_list + ["pop", "100%"]
print(second_list)


# Combining lists using .extend() method
# .extend() only takes one argument
# .extend() returns None
# It modifies the original list in place and returns nothing
# The extends unpacks the items

new_list = [1, 2, {"isActive": True}]

new_list.extend(["Bruce Lee", 99])
print(new_list)


# Adding new elements to a lists using append
# .append() adds one single item to the end of a list

my_list = [1, 2, 3]
my_list.append("Bruce Lee")
print(my_list)  # [1, 2, 3, "Bruce Lee"]

# Adding a nested list with append
my_list = [1, 2, 3]
my_list.append([4, 5])
print(my_list)  # [1, 2, 3, [4, 5]]  ← list inside a list!

# Modify a list
app_users = ["Ade", "Andrew", "Kingsley"]
app_users[0] = "Maxson"
print(app_users)

# Deleting elements of a list using del()
# Modifies the original list
del(app_users[0])
print(app_users)

# converting string to a list using split method
text = "Hello World!"
new_text = text.split()
print(new_text)

# Ex2
A = "The user: Users are awesome"
B = A.split(":")
print(B)


# Copy (copy by reference) the list A

A = ["hard rock", 10, 1.2]
B = A
print('A:', A)
print('B:', B)

# Copying a list using the .copy() method
#The `copy()` method is used to create a shallow copy of a list.

my_list = [1, 2, 3, 4, 5]
new_list = my_list.copy() 
print(new_list)


# Count method
# The `count()` method is used to count the number of occurrences of a specific element in a list in Python.

my_list = [1, 2, 2, 3, 4, 2, 5, 2]
count = my_list.count(2) 
print(count)

# Insert method
# The `insert()` method is used to insert an element.
# list_name.insert(index, element)

my_list = [1, 2, 3, 4, 5]
my_list.insert(2, 6)
print(my_list)


# Pop method
# `pop()` method is another way to remove an element from a list in Python. It removes and returns the element at the specified index. If you don't provide an index to the `pop()` method, it will remove and return the last element of the list by default

my_list = [10, 20, 30, 40, 50]
removed_element = my_list.pop(2) # Removes and returns the element at index 2
print(removed_element)


my_list = [10, 20, 30, 40, 50]
removed_element = my_list.pop() # Removes and returns the last element
print(removed_element)
# Output: 50

print(my_list)


# remove() method
# To remove an element from a list. The `remove()` method removes the first occurrence of the specified value.

my_list = [10, 20, 30, 40, 50]
my_list.remove(30) # Removes the element 30
print(my_list)
# Output: [10, 20, 40, 50]


# Reverse
# The `reverse()` method is used to reverse the order of elements in a list
my_list = [1, 2, 3, 4, 5]
my_list.reverse() 
print(my_list)


# Slicing
# You can use slicing to access a range of elements from a list.

my_list = [1, 2, 3, 4, 5]
print(my_list[1:4])
# Output: [2, 3, 4] (elements from index 1 to 3)

print(my_list[:3])
# Output: [1, 2, 3] (elements from the beginning up to index 2)

print(my_list[2:])
# Output: [3, 4, 5] (elements from index 2 to the end)

print(my_list[::2])
# Output: [1, 3, 5] (every second element)

# Sort Method
# The `sort()` method is used to sort the elements of a list in ascending order. If you want to sort the list in descending order, you can pass the `reverse=True` argument to the `sort()` method.

my_list = [5, 2, 8, 1, 9]
my_list.sort()
print(my_list)


my_list = [5, 2, 8, 1, 9]
my_list.sort(reverse=True)
print(my_list)
# Output: [9, 8, 5, 2, 1]

```

```python
# Dictionaries


person = {
    "name": "John Doe",
    "location": "United Kingdom",
    "age": 33,
    "role": "Python Developer"
}

print(person)

# Accessing properties of a dictionary
print(person["name"])
print(person["role"])

# Accessing properties using .get() method
print(person.get("location"))

# Accessing properties using dict()
value = person["name"]
print(value)

# Adding a new property to a dictionary
person["isEmployed"] = True
print(person)

# Deleting a property and it's value from a dictionary using del
del(person["age"])
print(person)

# Check if a property is in a dictionary using in
is_age_in_person = "age" in person
print(is_age_in_person)

# Print all keys in a dictionary
all_keys = person.keys()
print(all_keys)

all_values = person.values()
print(all_values)

# extract the keys of a dictionary as a list
list_keys = list(person.keys())
print(list_keys)

# extract the keys of a dictionary as a tuple
tuple_keys = tuple(person.keys())
print(tuple_keys)
```

```python
# Sets

# Sets are a type of collection. This means that like lists and tuples, you can input different python types
# Unlike lists and tuples, they are unordered. This means that sets do not record elements position.

# Sets only have unique elements. 

set1 = {1, 2, 3, 4, 2, 3, 8, 1, 1, 1, 2, 2, 4}
print(set1)

# Converting a list to a set using the set() method
list1 = [1, 2, 3, "John Doe", "John Doe", "john doe", 2, 1]

to_set = set(list1)
print(to_set)

# Add items to a set.
# To add an item to a set, we use .add() method

set2 = {"John", "Doe"}
set2.add("Igwe")

print(set2)

# Remove items from a set
# To remove an item from a set, we use .remove() method
set2.remove("John")
print(set2)

# Verify or check that an item is in a set
isItemInSet = "Igwe" in set2
print(isItemInSet)

# Intersection of 2 sets
set3 = {1, 2, 3, 5, 7, 9}
print(set3)

set4 = {1, 3, 8, 9, 11}
print(set4)

joint_set = set3 & set4
print(joint_set)

# Or use the .intersection method to find the intersection of 2 sets
joint2 = set3.intersection(set4)
print(f"The intersection of set 3 and set 4 is: {joint2}")

# Find the difference in set1 but not in set2
diff = set3.difference(set4)
print(diff)

# Get the union of all sets using the .union() method
union_set = set3.union(set4)
print(union_set)

# Check if a set is a subset of another set using the .issubset() method
is_subset = set3.issubset(union_set)
print(is_subset)

# check if superset using the .issuperset() method
result = union_set.issuperset(set3)
print(result)
```
