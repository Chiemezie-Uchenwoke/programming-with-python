```python
# Escape sequences (\)
print("The user who is excellent in manual testing \ngave her remarks")

# Includes a tab(space)
print("The user who is excellent in manual testing \tgave her remarks")

# Raw String: The r allows the use of backwards slash. In Python, raw strings are a powerful tool for handling textual data, especially when dealing with escape characters. By prefixing a string literal with the letter ‘r’, Python treats the string as raw, meaning it interprets backslashes as literal characters rather than escape sequences.
print(r"The user who is excellent in manual testing \ gave her remarks")

# Double backslashes: backslashes need to be escaped with another backslash to be displayed correctly.
print("Hello \\ how are you? \\ Welcome")
```

```python
# String Methods
# (1) Upper
A = "I am a manual tester"
B = A.upper()
print(B)

# (2) Replace
Text = "Ade is excellent in Python"
Replaced_Text = Text.replace("Ade", "Chiemezie")
print(Replaced_Text)

# (3) find - Returns the first index that matches the search. If the text provided doesn't exist, it returns -1.
sentence = "Be excellent at what you do"
print(sentence.find("excellent"))

# (4) Lower
New_Text = "Python Programming is Awesome"
Lowered_New_Text = New_Text.lower()
print(Lowered_New_Text)

# (5) Title - Used to make a text appear in title case
Sentence = "easter is appraoching fast"
print(Sentence.title())

# (6) Capitalize - Used to capitalize the first letter of the first word only
t = "easter is appraoching fast"
print(t.capitalize())
```
