```python
# Range: Prints a list(array) of numbers
# With 1 argument, it prints numbers from 0 but excluded the number itself
# With 2 arguments, it prints numbers from the first but before the second
range1 = list(range(5))
print(range1)

range2 = list(range(3, 10))
print(range2)
```

```python
# Loops

colors = ["red", "orange", "blue", "green", "torque"]
for i in range(0, 5):
    print(i)
    if (i % 2 == 0):
        print("Even")
    else:
        print("Odd")
    colors[i] = "white"

print(colors)
```

```python
# For Loops

colors = ["red", "orange", "blue", "green", "torque"]

for color in colors:
    message = ""
    if (color == "blue" or color == "green"):
        message = "favorite color"
        print(f"The color {color} is one of my {message}")
    
```

```python
# enumarate() function
colors = ["red", "orange", "blue", "green", "torque"]
print(list(enumerate(colors)))

for i,color in enumerate(colors):
    print(i)
    print(color)

# enumerate() gives you both the index and the value at the same time as you loop through a list.
```

```python
# For Loop
colors = ["red", "green", "blue"]
for index, color in enumerate(colors):
    print(index, color)

# You Can Also Start the Count From 1
for index, color in enumerate(colors, start=1):
    print(index, color)
```

```python
# Using the length of the list
colors = ["red", "green", "blue"]

for i in range(len(colors)):
    print(colors[i])
```
